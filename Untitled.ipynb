## 2.	Generate a series of plots to describe the temporal pattern (year-to-year, monthly, and day-of-week) of the overall crime incidence aggregating from all geo-locations

```python
# Getting values for plots
dict_year={}
dict_month={}
dict_day={}
years=[]
months=[]
days=[]
format = '%Y-%m-%d %H:%M:%S'
for val in data['CrimeDate']:
    year = datetime.datetime.strptime(str(val), format).year
    month = datetime.datetime.strptime(str(val), format).month
    day = datetime.datetime.strptime(str(val), format).weekday()
    years.append(year)
    months.append(month)
    days.append(day)
    if year in dict_year:
        dict_year[year] +=1
    else:
        dict_year[year] =1
    if month in dict_month:
        dict_month[month]+=1
    else:
        dict_month[month] = 1
    if day in dict_day:
        dict_day[day]+=1
    else:
        dict_day[day] = 1
```

```python
data['year'] = years
data['month'] = months
data['day'] = days
```
